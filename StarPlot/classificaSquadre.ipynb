```
import pandas as pd
df=pd.read_csv('season-1819_csv.csv') 
```

```
df
```

```
def punti(matchs,where,opp):
  punti=0
  gol_fatti=0
  gol_subiti=0
  p_vinte=0
  p_pareggiate=0
  p_perse=0
  tiri_fatti=0
  tiri_subiti=0
  tiri_in_porta_fatti=0
  tiri_in_porta_subiti=0
  falli_fatti=0
  falli_subiti=0
  corner_fatti=0
  corner_subiti=0
  gialli_presi=0
  gialli_avversari=0
  rossi_presi=0
  rossi_avversari=0
  
  for index, row in matchs.iterrows():
    gol_fatti+=row['FT'+where+'G']
    gol_subiti+=row['FT'+opp+'G']
    tiri_fatti+=row[where+'S']
    tiri_subiti+=row[opp+'S']
    tiri_in_porta_fatti+=row[where+'ST']
    tiri_in_porta_subiti+=row[opp+'ST']
    falli_fatti+=row[where+'F']
    falli_subiti+=row[opp+'F']
    corner_fatti+=row[where+'C']
    corner_subiti+=row[opp+'C']
    gialli_presi+=row[where+'Y']
    gialli_avversari+=row[opp+'Y']
    rossi_presi+=row[where+'R']
    rossi_avversari+=row[opp+'R']
    
    if row['FTR']==where:
      punti+=3
      p_vinte+=1
    elif row['FTR']=='D':
      punti+=1
      p_pareggiate+=1
    else:
      p_perse+=1
   
  return ["",punti,gol_fatti,gol_subiti,p_vinte,p_pareggiate,p_perse,tiri_fatti,
          tiri_subiti,tiri_in_porta_fatti,tiri_in_porta_subiti, falli_fatti, falli_subiti,
          corner_fatti,corner_subiti,gialli_presi,gialli_avversari,rossi_presi,rossi_avversari]

def classifica(club):
  s_home=df[df["HomeTeam"]==club]
  punti_casa=punti(s_home,"H","A")
  s_away=df[df["AwayTeam"]==club]
  punti_trasferta=punti(s_away,"A","H")
  punti_casa[0]=club
  punti_trasferta[0]=club
  
  punti_totali=[club]
  for i in range(1,19):
    punti_totali.append(punti_casa[i]+punti_trasferta[i])
  
  lista=punti_totali+punti_casa[1:]+punti_trasferta[1:]
  return lista

classifica("Roma")
```

```
squadre=['Juventus','Napoli','Atalanta','Inter','Milan','Roma','Torino','Lazio','Sampdoria','Bologna','Sassuolo','Udinese','Spal','Parma','Cagliari','Fiorentina','Genoa','Empoli','Frosinone','Chievo']
datafram=[]

for row in squadre:
  datafram.append(classifica(row))
 
```

```
bobo=pd.DataFrame(datafram, columns=['Club','Punti','GolFatti','GolSubiti','PartiteVinte','PartitePareggiate',
                                     'PartitePerse','TiriFatti','TiriSubiti','TiriInPortaFatti','TiriInPortaSubiti', 
                                     'FalliFatti', 'FalliSubiti','CornerFatti','CornerSubiti','GialliPresi',
                                     'GialliAvversari','RossiPresi','RossiAvversari'
                                    ,'PuntiCasa','GolFattiCasa','GolSubitiCasa','PartiteVinteCasa','PartitePareggiateCasa',
                                     'PartitePerseCasa','TiriFattiCasa','TiriSubitiCasa','TiriInPortaFattiCasa','TiriInPortaSubitiCasa', 
                                     'FalliFattiCasa', 'FalliSubitiCasa','CornerFattiCasa','CornerSubitiCasa','GialliPresiCasa',
                                     'GialliAvversariCasa','RossiPresiCasa','RossiAvversariCasa'
                                    ,'PuntiTransferta','GolFattiTransferta','GolSubitiTransferta','PartiteVinteTransferta','PartitePareggiateTransferta',
                                     'PartitePerseTransferta','TiriFattiTransferta','TiriSubitiTransferta','TiriInPortaFattiTransferta','TiriInPortaSubitiTransferta', 
                                     'FalliFattiTransferta', 'FalliSubitiTransferta','CornerFattiTransferta','CornerSubitiTransferta','GialliPresiTransferta',
                                     'GialliAvversariTransferta','RossiPresiTransferta','RossiAvversariTransferta'])
bobo
```
